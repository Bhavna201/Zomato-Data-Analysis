```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
df = pd.read_csv('/zomato.csv.zip')

print("Shape of Dataset:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nDataset Info:")
print(df.info())

print("\nFirst 5 Rows:")
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())
df.drop_duplicates(inplace=True)

df['rate'] = df['rate'].astype(str)
df['rate'] = df['rate'].str.replace('/5', '', regex=False)
df['rate'] = df['rate'].replace('NEW', np.nan)
df['rate'] = df['rate'].replace('-', np.nan)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

df['approx_cost(for two people)'] = (
    df['approx_cost(for two people)']
    .astype(str)
    .str.replace(',', '')
)

df['approx_cost(for two people)'] = pd.to_numeric(
    df['approx_cost(for two people)'],
    errors='coerce'
)

print("\nSummary Statistics:")
print(df.describe())

plt.figure(figsize=(12,6))
df['location'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Restaurant Locations')
plt.xlabel('Location')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='online_order', data=df)
plt.title('Online Order Availability')
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='book_table', data=df)
plt.title('Table Booking Availability')
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['rate'], bins=20)
plt.title('Restaurant Ratings Distribution')
plt.xlabel('Rating')
plt.show()

plt.figure(figsize=(12,6))
df['rest_type'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Restaurant Types')
plt.xlabel('Restaurant Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12,6))
df['cuisines'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='votes', y='rate', data=df)
plt.title('Votes vs Rating')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='online_order', y='rate', data=df)
plt.title('Online Order vs Rating')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='book_table', y='rate', data=df)
plt.title('Table Booking vs Rating')
plt.show()

top_rated = df[['name', 'rate']].sort_values(
    by='rate',
    ascending=False
).head(10)

print("\nTop Rated Restaurants:")
print(top_rated)

plt.figure(figsize=(8,5))
sns.histplot(df['approx_cost(for two people)'], bins=30)
plt.title('Cost Distribution for Two People')
plt.xlabel('Cost')
plt.show()

numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()

print("\nPROJECT INSIGHTS")
print("1. Most restaurants are concentrated in popular city areas.")
print("2. Online ordering is available in many restaurants.")
print("3. Higher votes generally indicate better ratings.")
print("4. Certain cuisines dominate the market.")
print("5. Restaurant ratings mostly lie between 3.5 and 4.5.")
```
